```python
import pandas as pd
import numpy as np
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import logging
```

```python
# Read CVS file
df = pd.read_csv('wiki_movie_plots_deduped.csv')
```

```python
# drop plot&genre na
df2=df.dropna(subset=['Genre','Plot'])
df2.info()
```

```python
# only plot & genre
df3 = df2[["Plot","Genre"]]
```

```python
df3.info()
```

```python
# only train comedy, horror, action, thriller, romance
train_genre = ['comedy', 'horror', 'action', 'thriller', 'romance']
indexNames = []
for i in train_genre:
    index = df3[(df3['Genre'] == i)].index.tolist()
    indexNames = indexNames + index
df4 = df3.loc[indexNames]
df4
```

```python
df4['Plot'].head(10)
```

```python
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
model_args = ClassificationArgs(num_train_epochs=1)
```

```python
model = ClassificationModel(
    'bert',
    'bert-base-cased',
    num_labels=3,
    args=model_args,
    use_cuda=False
) 
```

```python
# only train action, thriller, romance
train_genre = ['action', 'thriller', 'romance']
indexNames = []
for i in train_genre:
    index = df3[(df3['Genre'] == i)].index.tolist()
    indexNames = indexNames + index
df5 = df3.loc[indexNames]
df5
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df5['Genre'] = labelencoder.fit_transform(df5['Genre'])
df5
```

```python
df5["Genre"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df5, random_state=777, train_size=0.8)
print(train_df)
print(test_df)
```

```python
model.train_model(train_df)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(test_df)
```
